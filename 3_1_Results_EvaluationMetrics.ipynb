## 3.1.3 Merge Evaluation Metrics

```python
#Set new results dataframe
```

```python
#df_EvaluationMetrics = df_1
```

```python
EvaluationMetrics = [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12, df_13, df_14, df_15, df_16, df_17, df_18, df_19]
```

```python
df_EvaluationMetrics = pd.concat(EvaluationMetrics, ignore_index=True, keys=['df_1', 'df_2', 'df_3', 'df_4', 'df_5', 'df_6', 'df_7', 'df_8', 'df_9', 'df_10', 'df_11', 'df_12', 'df_13', 'df_14', 'df_15', 'df_16', 'df_17', 'df_18', 'df_19'], sort=False)
```

```python
#df_EvaluationMetrics = df_2.join(df_EvaluationMetrics.set_index('Model Run'), on='Model Run')
```

```python
#df_EvaluationMetrics = df_EvaluationMetrics.sort_values('Model Run')
```

```python
df_EvaluationMetrics
```

A scatter plot is generated to visually check the data is as expected.

```python
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
plt.scatter(df_EvaluationMetrics.R2_score, df_EvaluationMetrics.MAE, label=df_EvaluationMetrics.Model)
plt.xlabel("R2 score")
plt.ylabel("MAE")
plt.title("Evaluation Metrics")
plt.legend(loc=4)
plt.show()
```

```python
Models = ['GradientBoosting', 'RandomForest',  'Simple Linear', 'Multivariable Linear', 'Lasso', 'kNN', 'DecisionTree',]
```

```python
seaborn.set(style='darkgrid', font_scale=2)

fg = seaborn.FacetGrid(data=df_EvaluationMetrics, hue='Model', hue_order=Models, size=12, aspect=1.61)
fg.map(plt.scatter, 'R2_score', 'MAE').add_legend()
```

Saves the data frame to the folder '2_ModelResults' as a csv.

```python
df_EvaluationMetrics.to_csv('2_ModelResults/CombinedMetrics/EvaluationMetrics.csv')
```
